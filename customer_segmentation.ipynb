```python
# customer_segmentation.ipynb

# 1️⃣ Imports
from customer_segmentation import load_data, preprocess_data, apply_kmeans, plot_2d, plot_3d, cluster_summary
import plotly.io as pio
import sys

# 2️⃣ Set renderer for Jupyter Notebook
pio.renderers.default = "notebook_connected"
if "ipykernel" not in sys.modules:
    pio.renderers.default = "browser"

# 3️⃣ Load & preprocess
df = load_data()
X_scaled, features = preprocess_data(df)

# 4️⃣ Apply KMeans
df, kmeans = apply_kmeans(X_scaled, df, k=5)

# 5️⃣ Visualizations
plot_2d(df)   # Shows inline automatically
plot_3d(df)   # Should now render inline in notebook

# 6️⃣ Summary
cluster_summary(df, features)
```
